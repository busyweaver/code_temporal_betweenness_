```python
def bernoulli(p):
    x = randint(0,10^6)
    if x < p*10^6:
        return 1
    else:
        return 0
#change the set V for large graphs
def erdos_renyi_temp(n,tm,p, dire = True):
    g = set()
    V = list( [str(i) for i in range(n)] )
    for t in range(1,tm+1):
        for (a,b) in Combinations(V,2):
            if bernoulli(p):
                if bernoulli(0.5):
                    g.add((a,b,str(t)))
                    if not dire:
                        g.add((b,a,str(t)))
                        
                else:
                    g.add((b,a,str(t)))
                    if not dire:
                        g.add((a,b,str(t)))
                        
                    
    return g    
def to_csv(s,g):
    with open(s,"w") as f:
        for e in g:
            f.write( e[0]+" "+e[1]+" "+e[2]+"\n" )
```

```python
import subprocess
path = "../build.build/"
import time
# N = 40
# T = 12
# n = 20
# p = 1/(n*T)
def check_time_synth(N,T,n,p,c,typ, which):
    tot = 0
    tot2 = 0
    tot3 = 0
    for _ in range(N):
        g = erdos_renyi_temp(n,T,p)
        print("nb edges", len(g), "nTlog(nT)", n*T*math.log(n*T))
        to_csv('tmp.csv',g)
        if "dij" in which:
            st = time.time()
            subprocess.run([path+"btwBenchmark", "-f", "tmp.csv", "-d","-c",c,"-y",typ, "-o"]) 
            ed = time.time()
            tot2 += (ed - st)
        if "bell" in which:
            st = time.time()
            subprocess.run([path+"btwBenchmark", "-f", "tmp.csv", "-d","-c",c,"-y",typ,"-z","-o"]) 
            ed = time.time()
            tot += (ed - st)
        if "bfs" in which:
            st = time.time()
            subprocess.run([path+"btwBenchmark", "-f", "tmp.csv", "-d","-c",c,"-y",typ, "-b", "-o"]) 
            ed = time.time()
            tot3 += (ed - st)
#     print("average dijkstra ", tot/(N*2*2), "average bellman ", tot2/(N*2*2))
#     print("gain", tot/tot2)
    return tot, tot2, tot3
                
# x = check_time_synth(N,T,n,p)
```

```python
#varies nodes small number events
l = []
N = 450
gap = 50
which = ["bell", "dij", "bfs"]
for n in range(4,N+1,gap):
    print(n, " ", end = "")
    T = int(math.sqrt(n))
    p = 2/(T*n)
    repet = 1
    l.append( (n,check_time_synth(repet,T,n,p,"shortest","passive",which)) )
```

```python
#varies nodes small number events
l2 = []
which = ["bell", "dij", "bfs"]
for n in range(4,N+1,gap):
    print(n, " ", end = "")
    T = int(math.sqrt(n))
    p = 2/(T*n)
    repet = 1
    l2.append( (n,check_time_synth(repet,T,n,p,"shortestfastest","passive",which)) )
```

```python
import matplotlib.pyplot as plt 
labels = which
fig, ax = plt.subplots(layout='constrained')

for i in range(0,len(which),1):
    print( labels[i]+"_sh")
    plt.scatter(list(map(lambda x : x[0], l)), list(map(lambda x : x[1][i], l)), label = labels[i]+"_sh" )
for i in range(0,len(which)-1,1):
    print(labels[i]+"_shf")
    plt.scatter(list(map(lambda x : x[0], l2)), list(map(lambda x : x[1][i], l2)), label = labels[i]+"_shf" )

plt.legend()
ax.set_yscale('log')
ax.set_ylabel('Time in sec')
ax.set_xlabel('# nodes')
plt.savefig('sh_n.png', dpi=300)

plt.show() 
```
